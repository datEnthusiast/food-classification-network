```python
# list of classes
classes = labels.label.unique()

# file names in training directory
file_names = os.listdir('../data/train_set/train_set')

# define relative path
path = '../data/train_set_struc/'

# make directories
directories = [path + str(w) for w in classes]

for directory in directories:
    if not os.path.exists(directory):
        os.makedirs(directory)
```

```python
from shutil import copyfile

for file in file_names:
    row = labels[labels['img_name'].str.match(file)] # return row of match
    
    directory = str(row.label.values[0]) + '/' # corresponding label directory
    
    file_path = path + directory + file # generate target file path
    
    copyfile('../data/train_set/train_set/' + file, file_path) # copy file inside correct directory        
```
